#Absorció espectre EM visible per part de l'aigua

```
import matplotlib.pyplot as plt
import numpy as np

#data to plot
wvlen = []
a_coef = []
sigma = []
percent = []
f = open('dades_ab_em_h2o.txt', 'r') #tinc el fitxer en local
content = f.readlines()
for x in content:
    row = x.split()
    wvlen.append(float(row[0]))
    a_coef.append(float(row[1]))
    sigma.append(float(row[2]))
    percent.append(float(row[3]))

fig = plt.figure(figsize=(6.5,6))
ax = fig.add_subplot(111)
ax.set_yscale('log')
ax.set(xlabel=r"$\lambda$ (nm)", ylabel="Coef. d'absorció (m$^{-1}$)")
for tick in ax.yaxis.get_major_ticks():
  tick.label.set_fontsize(8) 
plt.tick_params(axis='both', which='both', direction='in')
ax.scatter(wvlen,a_coef, color= 'teal', s=4, zorder=3)
for i in range(len(wvlen)):
  ax.errorbar(wvlen[i], a_coef[i], xerr=sigma[i], capsize=2,color='k', alpha =0.8,zorder=2)
ax.annotate("(8, 1)",
            xy=(376, 1.3*a_coef[0]), xycoords='data',
            xytext=(-10, 25), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3")
            )
ax.annotate("8",
            xy=(401, 1.2*a_coef[wvlen.index(400)]), xycoords='data',
            xytext=(-3, 25), weight = 'bold', textcoords='offset points',
            arrowprops=dict(arrowstyle="-|>",
                            connectionstyle="arc3")
            )
ax.annotate("(7, 1)",
            xy=(418, 1.1*a_coef[wvlen.index(417.5)]), xycoords='data',
            xytext=(-10, 25), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3")
            )
ax.annotate("7",
            xy=(449, 1.1*a_coef[wvlen.index(450)]), xycoords='data',
            xytext=(-3, 25), weight = 'bold', textcoords='offset points',
            arrowprops=dict(arrowstyle="-|>",
                            connectionstyle="arc3")
            )
ax.annotate("(6, 1)",
            xy=(474, 1.1*a_coef[wvlen.index(475)]), xycoords='data',
            xytext=(-10, 25), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3")
            )
ax.annotate("6",
            xy=(514, 1.1*a_coef[wvlen.index(515)]), xycoords='data',
            xytext=(-3, 25), weight = 'bold', textcoords='offset points',
            arrowprops=dict(arrowstyle="-|>",
                            connectionstyle="arc3")
            )
ax.annotate("(5, 1)",
            xy=(550, 1.1*a_coef[wvlen.index(550)]), xycoords='data',
            xytext=(-14, 25), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3")
            )
ax.annotate("5",
            xy=(605, 1.1*a_coef[wvlen.index(605)]), xycoords='data',
            xytext=(-3, 25), weight = 'bold', textcoords='offset points',
            arrowprops=dict(arrowstyle="-|>",
                            connectionstyle="arc3")
            )
ax.annotate("(4, 1)",
            xy=(662, 1.1*a_coef[wvlen.index(662.5)]), xycoords='data',
            xytext=(-13, 25), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3")
            )
plt.savefig('ab_em_aigua_sim', dpi = 1000)
```

##Imatge

```
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 7

#plane data
xx_pla, yy_pla = np.meshgrid(np.linspace(-1,1,20), np.linspace(-1,1,20))
z_pla = np.zeros((20,20))

#curves data
#E_x
k = 30
yi = np.linspace(-np.sqrt(3)/2,0,100)
zi = np.linspace(1/2,0,100)
xi = 0.2 * np.cos(k * (yi*np.sqrt(3)/2-zi/2))

zii = np.linspace(0,1/2,100)
yi2 = np.linspace(0,-np.sqrt(3)/2,100)
zi2 = zii +  0.2 * np.cos(k * (yi2*np.sqrt(3)/2-zii/2) + np.pi/2 )
xi2 = [0 for i in range(100)]


yr = np.linspace(0,np.sqrt(3)/2,100)
zr = np.linspace(0,1/2,100)
xr = 0.2 * np.cos(k * (yr*np.sqrt(3)/2+zr/2))

ztt = np.linspace(0,-np.sqrt(3)/2,100)
yt = np.linspace(0,1/2,100)
zt = ztt +  0.2 * np.cos(k * (yt/2-ztt*np.sqrt(3)/2) + np.pi/2 )
xt = [0 for i in range(100)]

zt2 = np.linspace(0,-np.sqrt(3)/2,100)
yt2 = np.linspace(0,1/2,100)
xt2 = 0.1 * np.cos(k * (yt2/2-zt2*np.sqrt(3)/2))
# Create the figure
fig = plt.figure(figsize=(5,4))

# Add an axes
ax = fig.add_subplot(111,projection='3d')
ax._facecolors2d = ax._facecolor
# plot the surface
surf = ax.plot_surface(xx_pla, yy_pla, z_pla, alpha=0.5, color='silver', label='Interfície', zorder = 3)
surf._facecolors2d=surf._facecolors3d
surf._edgecolors2d=surf._edgecolors3d
# and plot the curves
ris = ax.plot(xi,yi,zi, c='teal', label = 'Llum incident', zorder=4)

rip = ax.plot(xi2,yi2,zi2, c='teal', zorder=4)

rrs = ax.plot(xr,yr,zr, c='gold', label= 'Llum reflectida', zorder=5)

rts = ax.plot(xt,yt,zt, c='dodgerblue', label= 'Llum transmesa', zorder=1)

rtp = ax.plot(xt2,yt2,zt2, c='dodgerblue', zorder=1)

#angle 90

xan = [0 for i in range(5)]
yani = np.linspace(1/16,(1+np.sqrt(3))/16,5)
zani = np.linspace(-np.sqrt(3)/16, (1-np.sqrt(3))/16,5)
yans = np.linspace(np.sqrt(3)/16,(1+np.sqrt(3))/16,5)
zans = np.linspace(1/16,(1-np.sqrt(3))/16,5)
rani = ax.plot(xan, yani, zani, c='black', zorder=2)
rans = ax.plot(xan,yans,zans, c='black',zorder=2)
dot = ax.scatter(0,(1+np.sqrt(3))/32,(1-np.sqrt(3))/32, s=5, c='black', zorder=2)
#ax.set_xlabel(r'$x$', rotation = 0)

#ax.set_ylabel('$y$')

#ax.set_zlabel('Z')


# Remove gray panes and axis grid
ax.xaxis.pane.fill = False
ax.xaxis.pane.set_edgecolor('white')
ax.yaxis.pane.fill = False
ax.yaxis.pane.set_edgecolor('white')
ax.zaxis.pane.fill = False
ax.zaxis.pane.set_edgecolor('white')
ax.grid(False)
#set the point of view
ax.view_init(elev=20, azim = 60) #20 60
ax.dist=10.5

# Remove axis ticks
ax.w_zaxis.line.set_lw(0.)
ax.set_zticks([])
ax.w_xaxis.line.set_lw(0.)
ax.set_xticks([])
ax.w_yaxis.line.set_lw(0.)
ax.set_yticks([])

ax.legend(loc=(0.61,0.28))
plt.tight_layout()
plt.savefig('Brewster_sim', dpi=1200)
```
